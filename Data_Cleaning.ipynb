```python
data = pd.read_csv('Dataset.csv')
```

```python
data.describe(), data.info()
```

Drop unnecessary columns

```python
data = data.drop(columns=['Unnamed: 0'], errors='ignore', axis=1)
```

Handle missing values

```python
data['Name'] = data['Name'].fillna('Unknown')
data['Nationality'] = data['Nationality'].fillna('Unknown')
```

Transform text columns

Extract numerical values from 'AP' and 'RP' columns

```python
data['AP'] = data['AP'].str.extract('(\d+)').astype(float)
data['RP'] = data['RP'].str.extract('(\d+)').astype(float)
```

Standardize categorical columns

```python
data['Card'] = data['Card'].str.upper()
data['Gender'] = data['Gender'].map({'M': 'Male', 'F': 'Female'})
```

Filter or correct negative values in 'Points'

```python
data['Points'] = data['Points'].apply(lambda x: x if x >= 0 else None)
```

Add a 'Season' column based on the 'Month' column

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'
```

```python
data['Season'] = data['Month'].apply(get_season)
```

Check for and remove duplicates

```python
data = data.drop_duplicates()
```

```python
data
```

Dive Experience

```python
data['experience_dive'] = data.groupby('Name').cumcount() + 1
```

Discipline Experience

```python
data['experience_discipline'] = data.groupby(['Name', 'Discipline']).cumcount() + 1
```

```python
data.to_csv('Dataset.csv')
```
